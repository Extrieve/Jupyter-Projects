```python
import pandas as pd
import numpy as np
```

```python
# How to initialize a data frame locally (99.99% of the time we load from a csv, excel, db or api)
df = pd.DataFrame({
    'Population': [35.467, 63.951, 80.94 , 60.665, 127.061, 64.511, 318.523],
    'GDP': [
        1785387,
        2833687,
        3874437,
        2167744,
        4602367,
        2950039,
        17348075
    ],
    'Surface Area': [
        9984670,
        640679,
        357114,
        301336,
        377930,
        242495,
        9525067
    ],
    'HDI': [
        0.913,
        0.888,
        0.916,
        0.873,
        0.891,
        0.907,
        0.915
    ],
    'Continent': [
        'America',
        'Europe',
        'Europe',
        'Europe',
        'Asia',
        'Europe',
        'America'
    ]
}, columns=['Population', 'GDP', 'Surface Area', 'HDI', 'Continent'])
```

## Indexing, Selection and Slicing

```python
df
```

```python
# Return a row by index
df.loc['Canada']
```

```python
# Return a row by numeric index
df.iloc[0]
```

```python
df['Population']
```

```python
# Maintain the DF look whilst selecting a column of data
df['Population'].to_frame()
```

```python
# Non inclusive range, unlike numpy
df[1:3]
```

```python
# More indexing scenarios
df.loc['France':'Italy']
```

```python
df.loc['France':'Italy', 'Population':'Surface Area']
```

```python
# Selecting with int indexes
df.iloc[1].to_frame()
```

```python
# Selecting multiple indexes
df.iloc[[0, 1, -1]]
```

```python
# Reverse the DataFrame's rows and columns
df.iloc[::-1, ::-1]
```

## Conditional Selection (boolean arrays)

```python
df
```

```python
df['Population'] > 70
```

```python
df.loc[df['Population'] > 70]
```

```python
# Conditional Statements in which we only return 1 row
df.loc[df['Population'] > 70, 'Population'].to_frame()
```

```python
# More stuff with conditional statements
df.loc[df['Population'] > 70, ['Population', 'GDP']]
```

## Data Drop / Removal

```python
df.drop('Canada')
```

```python
# Just showing that it is not a in place drop, you would have to reassingn the DataFrame in order to keep the changes.
df
```

```python
df.drop(['Canada', 'Japan'])
```

```python
# Dropping columns
df.drop(columns=['Population', 'HDI'])
```

```python
# Specify the axis in case some col name = row name
df.drop(['Italy', 'Canada'], axis=0)
```

```python
df.drop(['Population', 'HDI'], axis=1)
```

```python
df.drop(['Population', 'HDI'], axis=1)
```

```python
# Specifying Columns also works
df.drop(['Population', 'HDI'], axis='columns')
```

```python
df.drop(['Canada', 'Germany'], axis='rows')
```

## Modifying DataFrames

```python
# Adding new columns
langs = pd.Series(
    ['French', 'German', 'Italian'],
    index=['France', 'Germany', 'Italy'],
    name='Language'
)
```

```python
langs.to_frame()
```

```python
df['Language'] = langs
df
```

```python
df.drop('Canada, Language')
```

### Replacing Values per Columns

```python
df['Language'] = 'English'
```

```python
df
```

```python
df['Language'] = 'English'
df
```

### Renaming Columns

```python
df.rename(
    columns={
        'HDI': 'Human Development Index',
        'Anual Popcorn Consumption': 'APC'
    }, index={
        'United States': 'USA',
        'United Kingdom': 'UK',
        'Argentina': 'AR'
    })
```

```python
df = df.rename(index=str.upper)
df
```

```python
df = df.rename(index=lambda x: x.lower())
df
```

```python
for i in df.index:
    i = i[0].upper() + i[1:]
    print(i)
df
```

```python
# Dropping a column in place
df.drop('Language', inplace=True, axis=1)
df
```

### Adding Values

```python
df.append(pd.Series({
    'Population': 3,
    'GDP': 5
}, name='China'))
```

```python
df.loc['China'] = pd.Series({'Population': 1_400_000_000, 'Continent': 'Asia'})
```

```python
df.reset_index()
```

```python
df.set_index('Population')
```
