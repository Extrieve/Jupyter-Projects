```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Data/Restaraunt_cuisine.csv')
df
```

```python
cuisines = """French
Japanese
Seafood
Chinese
Asian
European
Filipino
American
Korean
Cafe
Italian
FastFood
Bakery
Brazilian
Pizza
Arabian
BarFood
Mexican
International
Peruvian
Desserts
Juices
Beverages
Lebanese
Burger
Steak
Indian
Sushi
BBQ
GourmetFastFood"""
cuisines = cuisines.split('\n')
cuisines
```

```python
dictCuisines = dict()

for i, cuisine in enumerate(cuisines):
    dictCuisines[cuisine] = i+1
dictCuisines
```

```python
df.head()
```

```python
df.loc[df['CUISINE_NAME'].str.contains('French'), 'CUISINE_ID'] = 1
df
```

```python
for item in dictCuisines.keys():
    df.loc[df['CUISINE_NAME'].str.contains(item), 'CUISINE_ID'] = dictCuisines[item]
```

```python
df
```

```python
df.to_csv('mapped_cuisines.csv', index=False)
```
