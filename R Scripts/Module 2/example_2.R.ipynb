```r
my.norms=rnorm(5000,72,14)
mean(my.norms)
hist(my.norms,col="red",main="My Little Red Histogram")
```

```r
sd(my.norms)
```

```r
# Density Plot
plot(density(my.norms),lwd=3,main="My Little Black Density Plot")
```

```r
# Histogram and Density Plot Together
hist(my.norms,col="red",main="My Histogram and Density Plot",
     ylim=c(0,.03),
     probability = TRUE)
lines(density(my.norms),lwd=3,col="blue")
```

```r
# QQ Plot
qqnorm(my.norms,pch=19,
       main="My QQ Plot")
qqline(my.norms,col="red",lwd=3)
```

```r
# My Distribution v. Theory
plot(density(my.norms),
     main="Comparing My Data and Theory",
     col="red",lwd=3,ylim=c(0,.03))
curve(dnorm(x,72,14),from=min(my.norms),
      to=max(my.norms),
      lwd=3,col="blue",add=TRUE)
```

```r
# Comparing Standard Deviations
curve(dnorm(x,72,14),from=30,to=120,lwd=3,ylim=c(0,.05),
      main="Comparing Standard Deviations")
for(i in 8:13){
  curve(dnorm(x,72,i),from=30,to=120,lwd=3,add=TRUE, col=i-7)
}
```

```r
# What About Uniform Distributions?
my.uniforms=runif(10000,0,1)
hist(my.uniforms, col="red") 
```

```r
plot(density(my.uniforms),lwd=3)
```

```r
qqnorm(my.uniforms,pch=19)
qqline(my.uniforms,col="red",lwd=3)
```

```r
# What About Objective Measures?
moments::skewness(my.uniforms)
moments::kurtosis(my.uniforms)
```
